# RandomForest Classifier

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as LR
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.metrics import accuracy_score
```

```python
!dir inputs
```

```python
train = pd.read_csv('inputs/train.tsv', sep='\t', index_col=0)
```

```python
train
```

```python
X = train.drop('class', axis=1)
y = train['class']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train
```

```python
y_train
```

```python
rf_model = RF()
rf_model.fit(X_train, y_train)
```

```python
predict = rf_model.predict(X_test)
len(predict)
```

```python
y_test
```

```python
accuracy_score(y_test, predict)
```

```python
test = pd.read_csv('inputs/test.tsv', sep='\t', index_col=0)
test
```

```python
predict_test = rf_model.predict(test)
```

```python
submit = pd.read_csv('inputs/sample_submit.csv', header=None)
submit
```

```python
submit[1] = predict_test
```

```python
submit
```

```python
submit.to_csv('outputs/20211103-001.csv', index=None, header=None)
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

```python
gscv = GridSearchCV(rf_model, param_grid)
```

```python
gscv.fit(X, y)
```

```python
gscv.best_params_
```

```python
predict_test = gscv.predict(test)
predict_test
```

```python
submit[1] = predict_test
submit
```

```python
submit.to_csv('outputs/20211103-002.csv', index=None, header=None)
```
