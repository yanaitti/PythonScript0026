```python
# 後ろ2文字を削除、float型に変換する関数change_area()を作成してください。
def change_area(x):
    return float(x[:-2])
    
# 面積カラムにchange_area()を適用させてください。
data['面積'] = data['面積'].apply(change_area)

# 面積カラムが正しく修正されたか確認してください。
data['面積'].head()
```

```python
# 新築なら0をそれ以外なら、年数を数値型で返す関数change_age()を作成してください。
def change_age(x):
    if x == '新築':
        return 0
    else:
        return int(x.split('年')[0])
        
# 築年数カラムを年数に修正してください。
data['築年数'] = data['築年数'].apply(change_age)

# 築年数カラムが正しく修正されたか確認。
print(data['築年数'].head())
```

```python
print(data['所在階'][20:40])
```

```python
print(data['所在階'])

# 部屋の階数を抜き出し、int型に変換する関数get_floor()を作成してください。
def get_floor(x):
    # (／を含まない)または(1文字目が／)なら欠損値を返す
    if ('／' not in x) or (x[0]=='／'):
        return np.nan
    else:
        floor = x[0:x.find('階')]
        if(floor.find('地下') > -1):
            return int(floor[floor.find('地下')+2:])
        else:
            return int(floor)
        
# 階数カラムを作成してください。
data['階数'] = data["所在階"].apply(get_floor)
# 階数カラムが正しく作成されたか確認。
print(data['階数'].unique())
print(data['階数'])

# 建物の階数を抜き出し、int型に変換する関数get_top_floor()を作成してください。
def get_top_floor(x):
    return int(x[x.find('／')+1:x.find('階建')])
    
# 最上階カラムを作成してください。
data['最上階'] = data['所在階'].apply(get_top_floor)
# 最上階カラムが正しく作成されたか確認。
print(data['最上階'].unique())
print(data['最上階'])
```

```python
import re

# 最も近い駅までの所要時間を返す関数get_min_time()を作成し、所要時間カラムを作成してください。
def get_min_time(x):
    time_list = re.findall(r'徒歩(\d+)分',str(x))
    
    # time_listを数値型に変換
    time_list = map(int, time_list)
    
    # time_listの最小値をmin_valueに代入
    min_value = min(time_list)
    
    return min_value

data['所要時間'] = data['アクセス'].apply(get_min_time)
# 所要時間カラムが正しく作成されたか確認。
print(data['所要時間'].head())
```

```python
import folium

def get_geojson_grid(upper_right, lower_left, n):

    all_boxes = []

    lat_steps = np.linspace(lower_left[0], upper_right[0], n+1)
    lon_steps = np.linspace(lower_left[1], upper_right[1], n+1)

    lat_stride = lat_steps[1] - lat_steps[0]
    lon_stride = lon_steps[1] - lon_steps[0]

    for lat in lat_steps[:-1]:
        for lon in lon_steps[:-1]:
            upper_left = [lon, lat + lat_stride]
            upper_right = [lon + lon_stride, lat + lat_stride]
            lower_right = [lon + lon_stride, lat]
            lower_left = [lon, lat]

            coordinates = [
                upper_left,
                upper_right,
                lower_right,
                lower_left,
                upper_left
            ]

            geo_json = {"type": "FeatureCollection",
                        "properties":{
                            "lower_left": lower_left,
                            "upper_right": upper_right
                        },
                        "features":[]}

            grid_feature = {
                "type":"Feature",
                "geometry":{
                    "type":"Polygon",
                    "coordinates": [coordinates],
                }
            }

            geo_json["features"].append(grid_feature)

            all_boxes.append(geo_json)

    return all_boxes

# 地図を作成
m = folium.Map(location=[35, 139])
print(data)

# ①グリッドを作成
min_lat = data['latitude'].min()
min_lon = data['longitude'].min()
max_lat = data['latitude'].max()
max_lon = data['longitude'].max()
grid = get_geojson_grid(upper_right=[max_lat, max_lon], lower_left=[min_lat,min_lon] , n=30)
```
